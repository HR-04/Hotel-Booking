```python
import pandas as pd

# Load cleaned data
df = pd.read_csv("hotel_bookings_cleaned.csv")

# Stratify by key columns (adjust weights if needed)
sampled_df = df.groupby(["is_canceled", "hotel"], group_keys=False)\
               .apply(lambda x: x.sample(frac=5000/len(df), random_state=42))

# Save sampled data
sampled_df.to_csv("hotel_bookings_sampled_5k.csv", index=False)
```
